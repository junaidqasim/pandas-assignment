```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
data.head()
```

```python
data.sort_values(["HP","Attack","Defense"])
```

```python
  Speed_mean = data["Speed"].mean()
    
def set_Speed(val):
    if val < Speed_mean:
        return "Speed Low"
    elif val == Speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"

```

```python
# "Speed_high_low"  -> "Speed High" , "Speed Low"
data["Speed_high_low"] = data["Speed"].apply(set_Speed)
data
```

```python
  HP_mean = data["HP"].mean()
     
def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    elif val == HP_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
# "HP_high_low"  -> "HP High" , "HP Low"
data["HP_high_low"] = data["HP"].apply(set_HP)
data
```
